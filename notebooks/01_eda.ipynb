```python
import pandas as pd
import matplotlib.pyplot as plt

# Daten laden
df = pd.read_csv("../data/D2_25-26.csv")
print(df.head())

# Beispiel: Verteilung Heim- vs. Auswärtstore
df[['FTHG','FTAG']].plot(kind="box")
plt.title("Heim- vs. Auswärtstore in Bundesliga 23/24")
plt.show()

# Durchschnittliche Punkte Heim vs. Auswärts
df['HomePoints'] = df['FTR'].map({'H':3, 'D':1, 'A':0})
df['AwayPoints'] = df['FTR'].map({'H':0, 'D':1, 'A':3})

print("Ø Heim-Punkte:", df['HomePoints'].mean())
print("Ø Auswärts-Punkte:", df['AwayPoints'].mean())

```
